```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Assuming your CSV file is located at /content/drive/MyDrive/world_cup_score.csv
file_path = "/content/drive/MyDrive/world_cup_score.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)


```

```python
import pandas as pd

# Assuming df is your DataFrame
print(df.columns)
```

```python
df.describe()
```

```python
# @title Year

from matplotlib import pyplot as plt
df['Year'].plot(kind='hist', bins=20, title='Year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Total Score for Team 1

from matplotlib import pyplot as plt
df['Total Score for Team 1'].plot(kind='line', figsize=(8, 4), title='Total Score for Team 1')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title Total Score for Team 2

from matplotlib import pyplot as plt
df['Total Score for Team 2'].plot(kind='line', figsize=(8, 4), title='Total Score for Team 2')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python


# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()





```

```python
# Distplot
plt.figure(figsize=(8, 6))
sns.histplot(df['Total Score for Team 1'], kde=True, color='blue', bins=20)
plt.title('Distribution of Total Score for Team 1')
plt.xlabel('Total Score for Team 1')
plt.ylabel('Probability Density Function')
plt.show()
```

```python
# Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Total Score for Team 1', y='Total Score for Team 2')
plt.title('Scatterplot of Total Scores for Team 1 and Team 2')
plt.show()
```

```python
# Lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Total Score for Team 1', marker='o', markersize=5, color='green')
sns.lineplot(data=df, x='Date', y='Total Score for Team 2', marker='o', markersize=5, color='blue')
plt.title('Total Score for Team 1 and Team 2 Over Time')
plt.xlabel('Date')
plt.ylabel('Total Score')
plt.xticks(rotation=45)
plt.legend(['Team 1', 'Team 2'])
plt.tight_layout()
plt.show()
```

```python
# Scatterplot for Total Wickets for Team 1 and Team 2
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Total Wicket for Team 1', y='Total Wicket for Team 2')
plt.title('Scatterplot of Total Wickets for Team 1 and Team 2')
plt.show()

```

```python

# Distplot for Runs Scored in over
plt.figure(figsize=(8, 6))
sns.histplot(df['Runs Scored in over'], kde=True, color='red', bins=20)
plt.title('Distribution of Runs Scored in over')
plt.xlabel('Runs Scored in over')
plt.ylabel('Probability Density Function')
plt.show()

# Distplot for Runs Scored till that over
plt.figure(figsize=(8, 6))
sns.histplot(df['Runs Scored till that over'], kde=True, color='purple', bins=20)
plt.title('Distribution of Runs Scored till that over')
plt.xlabel('Runs Scored till that over')
plt.ylabel('Probability Density Function')
plt.show()
```

```python
# Boxplot for Total Score for Team 1 and Team 2
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Total Score for Team 1', 'Total Score for Team 2']])
plt.title('Boxplot of Total Score for Team 1 and Team 2')
plt.xlabel('Team')
plt.ylabel('Total Score')
plt.show()





```

```python
# Selecting columns for the pairplot
selected_columns = ['Total Score for Team 1', 'Total Score for Team 2',
                    'Total Wicket for Team 1', 'Total Wicket for Team 2']

# Create pairplot with selected columns
sns.pairplot(df[selected_columns])
plt.title('Pairplot of Selected Columns')
plt.show()
```

```python
# Lineplot for Runs Scored in over and Runs Scored till that over Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Runs Scored in over', marker='o', markersize=5, color='green')
sns.lineplot(data=df, x='Date', y='Runs Scored till that over', marker='o', markersize=5, color='blue')
plt.title('Runs Scored in Over and Runs Scored till that over Over Time')
plt.xlabel('Date')
plt.ylabel('Runs Scored')
plt.xticks(rotation=45)
plt.legend(['Runs Scored in over', 'Runs Scored till that over'])
plt.tight_layout()
plt.show()
```
